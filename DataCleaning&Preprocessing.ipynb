## Data Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt
housing.hist(bins=50, figsize=(20,15), color='red')
```

```python
import seaborn as sns
figsize = (12, 1.2 * len(housing['Type'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(housing, x='Area (sqft)', y='Type', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
from pandas.plotting import scatter_matrix
attributes = ['Rent Price', 'Number of Rooms', 'Area (sqft)','Age (years)']
scatter_matrix(housing[attributes], figsize=(12,8))
```

```python
housing.plot(kind="scatter",x="Rent Price",y="Area (sqft)", alpha=0.15, title="Rent Price vs Area (sqft)")
```

```python
housing.plot(kind="scatter",x="Rent Price",y="Age (years)", alpha=0.15, title="Rent Price vs Age (years)")
```

```python
housing.plot(kind="scatter",x="Rent Price",y="Crime Rate", alpha=0.15, title="Rent Price vs Crime Rate")
```

###Correlations (Strong Positive Correlations)

Showing how Rent Price depends on other attributes

```python
housing_numeric = housing.select_dtypes(include=['number'])
corr_matrix = housing_numeric.corr()
#print(corr_matrix)
corr_matrix['Rent Price'].sort_values(ascending=False)
```

## Missing Attributes

- To take care of missing attributes you've three options :
1. Get rid of the missing datapoints (column)
2. Get rid of the whole attribute (row)
3. Set the value to some value(0,mean or median)

```python
#1
#housing.dropna(subset=["Number of Rooms"])
#housing.shape
```

```python
#2
#housing.drop("Number of Rooms", axis=1).shape

#Note:Original housing data remains unchanged
```

```python
median = housing["Number of Rooms"].median()
median

#If any missing case fill it with median
housing["Number of Rooms"].fillna(median)
housing.shape
```
